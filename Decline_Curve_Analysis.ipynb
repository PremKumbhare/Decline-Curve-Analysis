```python
#Importing the Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Lets assume required data
t=np.arange(1,1000)   #day
q_i=5000 #bbl/days
d=.005 # per day
```


Exponential Decline Curve

```python
#Calculation of flowrate for exponential Decline
b=0
q_e=q_i*np.exp(-d*t)

#Cumulative production
q_e_cum=(q_i-q_e)/d
```

```python
#Creating dataframe
df=pd.DataFrame({"time":t,"q":q_e,"q_cum":q_e_cum})
df.head()
```

```python
#Plot
plt.figure(figsize=(15,5))
plt.plot(df["time"],df["q"],label="Flowrate vs time",lw=4)
plt.title("Plot for Exponential Decline")
plt.xlabel("Time (day)")
plt.ylabel("Flowrate (bbl/day) ")
plt.grid()
plt.legend()
```

Harmonic Decline Curve
( b = 1 )

```python
#Calculating the flowrate for harmonic decline
b=1
q_h=q_i/(1+d*t)

#Cumulative production
q_h_cum=np.log(q_i/q_h)*(q_i/d)
```

```python
#Creating datframe
df=pd.DataFrame({"time":t,"q":q_h,"q_cum":q_h_cum})
df.head()
```

```python
#Plotting the data
plt.figure(figsize=(15,5))
plt.plot(df["time"],df["q"],lw=4,label="Flowrate vs time",color="green")
plt.title("Plot For Harmonic Decline")
plt.xlabel("Time (day)")
plt.ylabel("Flowrate (bbl/day)")
plt.grid()
plt.legend()
plt.show()
```

Hyperbolic Decline Curve

```python
#Take b as 0.4
b=0.4


```

```python
#Calculating flow rate for hyperbolic decline
q_hyp=q_i/np.power((b*d*t+1),(1/b))
q_hyp_cum=(q_i**b)*(q_hyp**(1-b)-q_i**(1-b))/((b-1)*d)
```

```python
#Creating Dataframe
df=pd.DataFrame({"time":t,"q":q_hyp,"q_cum":q_hyp_cum})
df.head()
```

```python
#Plotting the data
plt.figure(figsize=(15,5))
plt.plot(df["time"],df["q"],lw=4,label="Flowrate vs time")
plt.title("Plot for Hyperbolic Decline")
plt.xlabel("Time (day)")
plt.ylabel("Flowrate (bbl/day)")
plt.grid()
plt.legend()
plt.show()
```

Now Plotting all 3 together

```python
#Plot on same graph
plt.style.use("classic")
plt.figure(figsize=(20,10))
plt.plot(t,q_e,color="red",label="Exponential Decline Curve",lw=5)
plt.plot(t,q_h,color="green",label="Harmonic Decline Curve",lw=5)
plt.plot(t,q_hyp,"b",label="Hyperbolic Decline Curve",lw=5)
plt.title("Decline Curve")
plt.xlabel("Time (day)")
plt.ylabel("Q (bbl/day)")
plt.legend()
plt.grid()
plt.legend()
```
